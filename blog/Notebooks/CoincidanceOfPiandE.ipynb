```python
def Delta(a,b,c):
    return np.abs(np.pi**a+np.pi**b-np.e**c)/np.e**c

N = 200
A = np.arange(N+1)
B = np.arange(N+1)
C = np.arange(N+1)
C_least = np.zeros((N+1,N+1))
D = np.zeros((N+1,N+1))

for a in A:
    for b in B:
        temp = np.zeros((len(C),1))
        for c in C:
            temp[c] = Delta(a,b,c) 
        D[a,b] = np.min(temp)
        C_least[a,b] = np.argmin(temp)
```

```python
plt.figure(figsize=(14,10))
p = plt.imshow(C_least,norm=matplotlib.colors.LogNorm(vmin=C_least.min(), vmax=(C_least.max())))
plt.colorbar(p)
plt.show()
```

```python
plt.figure(figsize=(14,10))
plt.title(r'|$\pi^a$+$\pi^b$-$e^c$|/$e^c$')
p = plt.imshow(D,norm=matplotlib.colors.LogNorm(vmin=D.min(), vmax=(D.max())))
plt.colorbar(p)
plt.show()
```
